```python
person_pg,person_dis,person_con,org_pg,org_dis,org_con,loc_pg,loc_dis,loc_con=[],[],[],[],[],[],[],[],[]
```

```python
for v in (norm_loc.keys()):
    print(v)
```

```python
%%time

for v in norm_loc.keys():  
    
    try:
        html = wiki.page(v).html()
    except wiki.DisambiguationError as e:
        loc_dis.append(v)
    except wiki.PageError:
        loc_pg.append(v)
    except ConnectionError:
        loc_con.append(v)
    except:
        pass
```

```python
import pickle
```

```python
   
with open("loc_dis.txt","wb") as fp:
    pickle.dump(loc_dis,fp)
with open("loc_pg.txt","wb") as fp:
    pickle.dump(loc_pg,fp)
with open("loc_con.txt","wb") as fp:
    pickle.dump(loc_con,fp)
```
