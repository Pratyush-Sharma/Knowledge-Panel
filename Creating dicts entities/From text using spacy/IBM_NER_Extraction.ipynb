```python
def clean(text):    
    
    text = text.replace('\n',' ')
    text = text.strip()    
    return text
```

```python
# A function to extract all the entities from a text and return a dictionary of each type of entities

def NER_extract(text):
    
    dic = {}
    text = clean(text)
    
    doc = nlp(text)
    
    for ent in doc.ents:
        dic[str(ent.label_)] = set()
    
    for ent in doc.ents:
        dic[str(ent.label_)].add(ent.text) 
    
    for key,value in dic.items():
        dic[key] = list(dic[key])
    
    return dic
```

```python
Dict_ls = []
offset = 0

while offset<=261000:
    count = 3000
    reqparams = {'offset':offset,'count':count}
    discovery_data = dh_help.getDiscoveryData(discovery_config,reqparams)
    
    if (offset+3000)<=261000:
        offset+=3000
    else
        offset = offset+(261000-offset)
    
    result = discovery_data['results']
    for obj in result:
        temp = {}
        temp['id'] = obj['id']
        temp = {**temp,**NER_extract(obj['text'])}
        Dict_ls.append(temp)
```
